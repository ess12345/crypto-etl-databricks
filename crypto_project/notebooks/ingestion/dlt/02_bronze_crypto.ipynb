```python
import dlt
from pyspark.sql.functions import current_timestamp

@dlt.table(
    name="bronze_crypto_market",
    comment="Refined Bronze table: API ingestion with metadata.",
    table_properties={"quality": "bronze"}
)
def bronze_crypto_market():

    df = (
        spark.readStream
            .format("cloudFiles")
            .option("cloudFiles.format", "json")
            .option("cloudFiles.inferColumnTypes", "true")
            .option("multiline", "true")  # required for JSON array format
            .load("/Volumes/crypto_lake/raw/raw_crypto/crypto/raw/")
    )

    return df.withColumn("ingestion_timestamp", current_timestamp())

```
